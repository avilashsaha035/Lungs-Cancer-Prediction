```python
import numpy as np
import pandas as pd
import seaborn
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split    #split dataset func

from sklearn.ensemble import RandomForestClassifier



# confusion matrix in sklearn
from sklearn.metrics import confusion_matrix, plot_roc_curve
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
 
```

```python
df = pd.read_csv("cancer_patient_data.csv")
```

```python
df.drop(["index","Patient Id"], axis=1, inplace = True)
```

```python
df.replace(to_replace = "Low" , value = 0 , inplace = True)
df.replace(to_replace = "Medium" , value = 1 , inplace = True)
df.replace(to_replace = "High" , value = 1 , inplace = True)
```

```python
data0 = df[df['Level'] == 0]
data1 = df[df['Level'] == 1]
total_0 = data0["Level"].count()
total_1 = data1["Level"].count()
print("Number of 0 data:",total_0)
print("Number of 1 data:",total_1)

pct_0 = total_0 / (total_0 + total_1) * 100
pct_1 = total_1 / (total_0 + total_1) * 100
print("Number of O data = ",pct_0,"%")
print("Number of 1 data = ",pct_1,"%")
```

```python
target=df['Level']
```

# spilt dataset into training and testing

```python
training, testing = train_test_split(df, test_size=0.3, shuffle=True, random_state=32, stratify=target)

print(pd.value_counts(training['Level']))
print(pd.value_counts(testing['Level']))
```

# Percantage of Training Data

```python
total_1 = training["Level"].sum()
total = training["Level"].count()
print("Number of 1 data:",(total_1/total)*100,"%")
```

# Percantage of Testing Data

```python
total_1 = testing["Level"].sum()
total = testing["Level"].count()
print("Number of 1 data:",(total_1/total)*100,"%")
```

# separate the target variables from training data

```python
train_x = training.drop(columns=['Level'],axis=1)
```

```python
train_y = training['Level']
```

# separate the target variables from test data

```python
test_x = testing.drop(columns=['Level'],axis=1)
```

```python
test_y = testing['Level']
```

# Building a model

```python
model = RandomForestClassifier()
model.fit(train_x , train_x)
```

```python
y_pred = model.predict(test_x)
```
